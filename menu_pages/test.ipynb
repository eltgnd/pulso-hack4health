```python
import pandas as pd
import numpy as np

# List of barangays
barangay_list = [
    "Aliwekwek", "Baay", "Balangobong", "Balococ", "Bantayan", "Basing",
    "Capandanan", "Domalandan Center", "Domalandan East", "Domalandan West",
    "Dorongan", "Dulag", "Estanza", "Lasip", "Libsong East", "Libsong West",
    "Malawa", "Malimpuec", "Maniboc", "Matalava", "Naguelguel", "Namolan",
    "Pangapisan North", "Pangapisan Sur", "Poblacion", "Quibaol", "Rosario",
    "Sabangan", "Talogtog", "Tonton", "Tumbar", "Wawa"
]

# Initialize a DataFrame to store the results
summary_df = pd.DataFrame(columns=[
    'Barangay', 'Population', 'Total Cases (2024)',
    'Previous 5-6 weeks', 'Previous 3-4 weeks', 'Recent 1-2 weeks',
    'Growth Rate 3-4 vs 5-6', 'Growth Rate 1-2 vs 3-4'
])

# Loop through each barangay
for barangay in barangay_list:
    # Load the CSV file for the current barangay
    file_path = f"../datasets/barangay_MRS_dataset/{barangay}_MRS_data.csv"
    df = pd.read_csv(file_path)
    
    # Calculate total cases for each period
    previous_5_6_weeks = df.loc[df['Week'].isin([5, 6]), '2024 Cases'].sum()
    previous_3_4_weeks = df.loc[df['Week'].isin([3, 4]), '2024 Cases'].sum()
    recent_1_2_weeks = df.loc[df['Week'].isin([1, 2]), '2024 Cases'].sum()
    
    # Calculate growth rates
    if previous_5_6_weeks != 0:
        growth_rate_3_4_vs_5_6 = (previous_3_4_weeks - previous_5_6_weeks) / previous_5_6_weeks * 100
    else:
        growth_rate_3_4_vs_5_6 = np.nan
    
    if previous_3_4_weeks != 0:
        growth_rate_1_2_vs_3_4 = (recent_1_2_weeks - previous_3_4_weeks) / previous_3_4_weeks * 100
    else:
        growth_rate_1_2_vs_3_4 = np.nan
    
    # Append the results to the summary DataFrame
    summary_df = summary_df.append({
        'Barangay': barangay,
        'Population': np.nan,  # Placeholder for population data
        'Total Cases (2024)': df['2024 Cases'].sum(),
        'Previous 5-6 weeks': previous_5_6_weeks,
        'Previous 3-4 weeks': previous_3_4_weeks,
        'Recent 1-2 weeks': recent_1_2_weeks,
        'Growth Rate 3-4 vs 5-6': growth_rate_3_4_vs_5_6,
        'Growth Rate 1-2 vs 3-4': growth_rate_1_2_vs_3_4
    }, ignore_index=True)

# Handle division by zero and infinite values
summary_df.replace([np.inf, -np.inf], np.nan, inplace=True)
summary_df.fillna(0, inplace=True)

# Save the result to a new CSV file
summary_df.to_csv('barangay_growth_rates_summary.csv', index=False)

print(summary_df)
```

```python
import pandas as pd
import numpy as np

# List of barangays
barangay_list = [
    "Aliwekwek", "Baay", "Balangobong", "Balococ", "Bantayan", "Basing",
    "Capandanan", "Domalandan Center", "Domalandan East", "Domalandan West",
    "Dorongan", "Dulag", "Estanza", "Lasip", "Libsong East", "Libsong West",
    "Malawa", "Malimpuec", "Maniboc", "Matalava", "Naguelguel", "Namolan",
    "Pangapisan North", "Pangapisan Sur", "Poblacion", "Quibaol", "Rosario",
    "Sabangan", "Talogtog", "Tonton", "Tumbar", "Wawa"
]

# Initialize a DataFrame to store the results
summary_df = pd.DataFrame(columns=[
    'Barangay', 'Population', 'Total Cases (2024)',
    'Previous 5-6 weeks', 'Previous 3-4 weeks', 'Recent 1-2 weeks',
    'Growth Rate 3-4 vs 5-6', 'Growth Rate 1-2 vs 3-4'
])

# Loop through each barangay
for barangay in barangay_list:
    # Load the CSV file for the current barangay
    file_path = f"../datasets/barangay_MRS_dataset/{barangay}_MRS_data.csv"
    df = pd.read_csv(file_path)
    
    # Calculate total cases for each period
    previous_5_6_weeks = df.loc[df['Week'].isin([5, 6]), '2024 Cases'].sum()
    previous_3_4_weeks = df.loc[df['Week'].isin([3, 4]), '2024 Cases'].sum()
    recent_1_2_weeks = df.loc[df['Week'].isin([1, 2]), '2024 Cases'].sum()
    
    # Calculate growth rates
    if previous_5_6_weeks != 0:
        growth_rate_3_4_vs_5_6 = (previous_3_4_weeks - previous_5_6_weeks) / previous_5_6_weeks * 100
    else:
        growth_rate_3_4_vs_5_6 = np.nan
    
    if previous_3_4_weeks != 0:
        growth_rate_1_2_vs_3_4 = (recent_1_2_weeks - previous_3_4_weeks) / previous_3_4_weeks * 100
    else:
        growth_rate_1_2_vs_3_4 = np.nan
    
    # Append the results to the summary DataFrame using pd.concat
    result = pd.DataFrame([{
        'Barangay': barangay,
        'Population': np.nan,  # Placeholder for population data
        'Total Cases (2024)': df['2024 Cases'].sum(),
        'Previous 5-6 weeks': previous_5_6_weeks,
        'Previous 3-4 weeks': previous_3_4_weeks,
        'Recent 1-2 weeks': recent_1_2_weeks,
        'Growth Rate 3-4 vs 5-6': growth_rate_3_4_vs_5_6,
        'Growth Rate 1-2 vs 3-4': growth_rate_1_2_vs_3_4
    }])
    summary_df = pd.concat([summary_df, result], ignore_index=True)

# Handle division by zero and infinite values
summary_df.replace([np.inf, -np.inf], np.nan, inplace=True)
summary_df.fillna(0, inplace=True)

# Save the result to a new CSV file
summary_df.to_csv('barangay_growth_rates_summary.csv', index=False)

print(summary_df)
```
